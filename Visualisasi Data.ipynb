```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca file CSV
df = pd.read_csv('cleaned_ifberita.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
df['Tanggal'].hist(bins=30, edgecolor='black')
plt.title('Distribution of Date')
plt.xlabel('Tanggal')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right') # Untuk memutar label sumbu x
plt.show()
```

```python
# Bar plot untuk kolom 'Kategori'
plt.figure(figsize=(10, 4))
df['Tanggal'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Count of Kategori')
plt.xlabel('Kategori')
plt.ylabel('Total')
plt.show()
```
