```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd 
import requests
import csv 
```

```python
def extract_info(url):
    req = Request(url)
    webpage = urlopen(req).read()
    
    with requests.Session() as c:
         soup = BeautifulSoup(webpage, 'html5lib')

    
    for item, ex, date in zip(soup.find_all('h2', attrs={'class': 'entry-title'}), 
                        soup.find_all('div', attrs={'class': 'above-entry-meta'}),
                        soup.find_all('time', attrs={'class': 'entry-date published'})):
        
        link = item.find('a', href=True)['href']
        title = item.find('a', href=True)['title']
        
        kat = ex.find_all('a')
        kate = [category.text.strip() for category in kat]
        date = date.text.strip()
        
        print("Judul:", title)
        print("Link:", link)
        print("Kategori:", kate)
        print("Tanggal:", date)
        print()
```

```python
link17 = "https://if.unila.ac.id/"
req = Request(link17)
webpage = urlopen(req).read()

with requests.Session() as c:
    soup = BeautifulSoup(webpage, 'html5lib')

csv_filename = "ifberita.csv"

with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Judul', 'Link', 'Kategori', 'Tanggal']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for item17, ex, date in zip(soup.find_all('h3', attrs={'class': 'entry-title'}), 
                                soup.find_all('div', attrs={'class': 'above-entry-meta'}),
                                soup.find_all('time', attrs={'class': 'entry-date published'})):
        link = item17.find('a', href=True)['href']
        title = item17.find('a', href=True)['title']

        kat = ex.find_all('a')
        kate = ', '.join([category.text.strip() for category in kat])
        date_text = date.text.strip()

        # Menulis ke file CSV
        writer.writerow({'Judul': title, 'Link': link, 'Kategori': kate, 'Tanggal': date_text})

        # Mencetak ke layar
        print("Judul:", title)
        print("Link:", link)
        print("Tanggal:", date_text)
        print("Kategori:")
        for category in kat:
            print("-", category.text.strip())
        print()
```
